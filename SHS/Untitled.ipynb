```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
swiss_im = Image.open("images/map/Suisse_summary.png")
plt.figure(figsize=(15, 10))
plt.imshow(swiss_im)
plt.axis("off")
plt.show()
```

```python
swiss_arr=np.array(swiss_im)
print(swiss_arr.shape)
```

```python
def toHex(rgba):
    return "#" + "".join(["{:0<2X}".format(x) for x in rgba])
```

```python
colors = {toHex(rgba): n for (n, rgba) in swiss_im.getcolors()}
```

```python
for c, n in colors.items():
    print(c, n)
```

```python
allSuperficy = sum([n for n in colors.values()])
swissSuperficy = sum([n for c, n in colors.items() if c not in ["#000000FF"]])
gt30percent = sum([n for c, n in colors.items() if c in ["#D55500FF"]])
inaccessible = sum([n for c, n in colors.items() if c in ["#D55500FF", "#FFAD77FF"]])

print("Superficy with slope > 30°: {:.2f}%\nSuperficy inccessible > 30°: {:.2f}%"\
      .format(gt30percent / swissSuperficy * 100,
             inaccessible / swissSuperficy * 100))
```
